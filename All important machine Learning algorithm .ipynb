```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns 
```

```python
pd.set_option("display.max_columns",None)
dataset=pd.read_csv("new_student2.csv")
dataset
```

```python
dataset.drop('school',inplace=True,axis=1)
```

```python
dataset.info()
```

```python
object_index=dataset.select_dtypes(include='object').columns
object_index
```

```python
object_dataset=pd.DataFrame(dataset,columns=['sex', 'address', 'famsize', 'Pstatus', 'Mjob', 'Fjob', 'reason',
       'guardian', 'schoolsup', 'famsup', 'paid', 'activities', 'nursery',
       'higher', 'internet', 'romantic'])
object_dataset
```

```python

new_object_dataset=pd.get_dummies(object_dataset,drop_first=True)
new_object_dataset
```

```python
new_dataset=dataset.drop(['sex', 'address', 'famsize', 'Pstatus', 'Mjob', 'Fjob', 'reason',
       'guardian', 'schoolsup', 'famsup', 'paid', 'activities', 'nursery',
       'higher', 'internet', 'romantic'],axis=1)
new_dataset
```

```python
new_=pd.concat([new_object_dataset,new_dataset],axis=1)
```

```python
X=new_.iloc[:,:-1]
Y=new_.iloc[:,-1:]
```

```python
#pass=1
#fail=0
y_array=np.array(Y)
y_array=y_array.ravel()
new_y=[]
for i in y_array:
    if i>9:
        new_y.append(1)
    else:
        new_y.append(0)
new_y=pd.DataFrame(new_y,columns=['G3'])
new_y
```

```python
from sklearn.metrics import recall_score, roc_auc_score, precision_score,f1_score
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,new_y,test_size=0.2)
```

# Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
Decion_tree=DecisionTreeClassifier()
Decion_tree.fit(X_train,y_train)
Decion_pred1=Decion_tree.predict(X_test)
```

```python
#Recall Score
recall_score(y_test,Decion_pred1)
```

```python
#AUC score
roc_auc_score(y_test,Decion_pred1)
```

```python
#Precision score
precision_score(y_test,Decion_pred1)
```

```python
#F1_score
f1_score(y_test,Decion_pred1)
```

```python
Random_forest.fit(X_train,y_train)
rand_pred=Random_forest.predict(X_test)
```
